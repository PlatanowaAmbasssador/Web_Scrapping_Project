```python
import pandas as pd
```

## Nowe auta

```python
# Load the dataset
df = pd.read_pickle("df_car_scrapped_data.pkl")
df
```

```python
# Convert 'Price' column
if 'Price' in df.columns:
    df['Price'] = df['Price'].str.replace(" zł", "", regex=False)
    df['Price'] = df['Price'].str.replace(" ", "", regex=True)
    df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Convert 'Rok Produkcji' (Year of Production)
if 'Rok Produkcji' in df.columns:
    df['Rok Produkcji'] = pd.to_numeric(df['Rok Produkcji'], errors='coerce')


# Convert 'Pojemnosc Silnika' (Engine Capacity)
if 'Pojemnosc Silnika' in df.columns:
    df['Pojemnosc Silnika'] = df['Pojemnosc Silnika'].str.replace(" cm3", "", regex=False)
    df['Pojemnosc Silnika'] = pd.to_numeric(df['Pojemnosc Silnika'], errors='coerce')

# Convert 'Moc Silnika' (Engine Power)
if 'Moc Silnika' in df.columns:
    df['Moc Silnika'] = df['Moc Silnika'].str.replace(" KM", "", regex=False)
    df['Moc Silnika'] = pd.to_numeric(df['Moc Silnika'], errors='coerce')

# Convert dimensions and weight columns
cols_to_convert = ['Dlugosc', 'Szerokosc', 'Wysokosc', 'Rozstaw Osi']
for col in cols_to_convert:
    if col in df.columns:
        df[col] = df[col].str.replace(" cm", "", regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert 'Masa Calkowita' (Total Weight)
if 'Masa Calkowita' in df.columns:
    df['Masa Calkowita'] = df['Masa Calkowita'].str.replace(" kg", "", regex=False)
    df['Masa Calkowita'] = pd.to_numeric(df['Masa Calkowita'], errors='coerce')
```

```python
# Check missing values
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

## Car grouping by price

```python
# Define price thresholds using quantiles
low_threshold = df_combined["Price"].quantile(0.33)
high_threshold = df_combined["Price"].quantile(0.66)

# Assign each car to a price group
def categorize_price(price):
    if price <= low_threshold:
        return "Low Price"
    elif price <= high_threshold:
        return "Mid Price"
    else:
        return "High Price"

df_combined["Price Group"] = df_combined["Price"].apply(categorize_price)

# Check distribution
df_combined["Price Group"].value_counts()
```

```python
# Boxplot: Engine Size vs. Price Group
plt.figure(figsize=(8, 5))
sns.boxplot(x="Price Group", y="Pojemnosc Silnika", data=df_combined)
plt.title("Engine Size Distribution by Price Group")
plt.show()

# Boxplot: Engine Power vs. Price Group
plt.figure(figsize=(8, 5))
sns.boxplot(x="Price Group", y="Moc Silnika", data=df_combined)
plt.title("Engine Power by Price Group")
plt.show()

# Count Plot: Drivetrain Type vs. Price Group
plt.figure(figsize=(8, 5))
sns.countplot(x="Naped", hue="Price Group", data=df_combined)
plt.title("Drivetrain Type by Price Group")
plt.xticks(rotation=45)
plt.show()

# Count Plot: Car Body Type vs. Price Group
plt.figure(figsize=(8, 5))
sns.countplot(x="Typ Nadwozia", hue="Price Group", data=df_combined)
plt.title("Car Body Type by Price Group")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Car Brand and compute mean price + count
brand_prices = df_combined.groupby("Car Model").agg(
    Avg_Price=("Price", "mean"),  # Average price per brand
    Count=("Car Model", "count")  # Number of cars per brand
).reset_index()

# Sort by average price (descending)
brand_prices = brand_prices.sort_values(by="Avg_Price", ascending=False)

# Show the result
print(brand_prices)

```

```python
# Group by Car Model and compute the average price
car_prices = df_combined.groupby("Car Model")["Price"].mean().reset_index()

# Sort by average price in descending order
car_prices = car_prices.sort_values(by="Price", ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x="Price", y="Car Model", data=car_prices, palette="viridis")

# Labels and title
plt.xlabel("Average Price (PLN)")
plt.ylabel("Car Model")
plt.title("Average Car Price by Model")
plt.show()
```
